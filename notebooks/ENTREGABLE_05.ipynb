CELDA 2: Crear el archivo main.py

```python
%%writefile main.py

from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware
from fastapi.responses import JSONResponse
import json

app = FastAPI(
    title="API Investing AI",
    description="API REST de los modelos financieros",
    version="1.0"
)

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=False,
    allow_methods=["*"],
    allow_headers=["*"],
)

@app.get("/")
def inicio():
    return {"mensaje": "API funcionando correctamente"}

@app.get("/mercado")
def mercado():
    return {
        "FSM": "OK",
        "BVN": "OK"
    }

@app.get("/svc")
def svc():
    with open("N2_svc.json", "r", encoding="utf-8") as f:
        datos = json.load(f)
    return JSONResponse(content=datos)

@app.get("/rnns")
def rnns():
    with open("N3_rnns.json", "r", encoding="utf-8") as f:
        datos = json.load(f)
    return JSONResponse(content=datos)

@app.get("/lstm")
def lstm():
    with open("N4_lstm_regressor.json", "r", encoding="utf-8") as f:
        datos = json.load(f)
    return JSONResponse(content=datos)
```
